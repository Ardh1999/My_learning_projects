```
import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```
img = image.load_img("/content/00000039.jpg", target_size=(300,300))
img = np.array(img).reshape(1,300,300,3)
```

```
img.shape
```

```
plt.imshow(img[0])
```

```
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    zoom_range = 0.3,
    shear_range =0.3,
    height_shift_range=0.2,
    horizontal_flip=True)
```

```
my_data = datagen.flow(x=img, batch_size=1)
c = 0
for i in my_data:
  if c==10:
    break

  print(i.shape)
  plt.imshow(i[0]/255.)
  plt.show()
  c+=1
```
