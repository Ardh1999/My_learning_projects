```
from keras.preprocessing import image
```

```
train_datagen = image.ImageDataGenerator(rotation_range=20,
                                         width_shift_range = 0.2,
                                         height_shift_range = 0.2,
                                         shear_range = 0.2,
                                         zoom_range=0.3,
                                         horizontal_flip = True
                                         )
```

```
val_datagen = image.ImageDataGenerator()
```

```
train_generator = train_datagen.flow_from_directory("images/",
                                  target_size=(224,224),  # image ka size
                                  class_mode='categorical',
                                  shuffle=True,
                                  batch_size=32 )
```

```
train_generator.class_indices
```

```
val_generator = val_datagen.flow_from_directory("val_images/", 
                                                target_size=(224,224), 
                                                class_mode='categorical', 
                                                batch_size=32)
```

```
from keras.applications.resnet50 import ResNet50
from keras.applications.resnet50 import preprocess_input,decode_predictions
```

```
resnet = ResNet50(include_top = False ,weights = "imagenet", input_shape=(224,224,3))
```

```
resnet.summary()
```

```
from keras.layers import Input, Conv2D, MaxPool2D, Dense, Flatten, GlobalAveragePooling2D
from keras.models import Model
```

```
gap = GlobalAveragePooling2D()(resnet.output)
fc1 = Dense(4, 'softmax')(gap)
```

```
my_model = Model(inputs = resnet.input  , outputs = fc1)
```

```
my_model.summary()
```

```
for layer in my_model.layers[0:-2]:
  layer.trainable = False
```

```
my_model.summary()
```

```
my_model.compile(optimizer='adam', loss="categorical_crossentropy",  metrics = ['acc'])
```

```
his = my_model.fit_generator(train_generator, 724//32, 10, validation_data=val_generator, validation_steps=81//32)
```
