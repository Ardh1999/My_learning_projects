```
from keras.applications.resnet50 import ResNet50
from keras.applications.resnet50 import preprocess_input, decode_predictions

import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing import image
from keras.layers import Dense
from keras.models import Model
```

```
resnet = ResNet50()
```

```
resnet.summary()
```

```
# resnet.trainable = False
```

```
for layer in resnet.layers[:-6]:
  layer.trainable = False
```

```
resnet.summary()
```

```
second_last_layer_output = resnet.layers[-2].output
```

```
fc1 = Dense(256, activation='relu')(second_last_layer_output)
fc2 = Dense(4, activation='softmax')(fc1)
```

```
my_model = Model(inputs = resnet.input, outputs = fc2 )
```

```
my_model.summary()
```

```
for layer in my_model.layers[0:-3]:
  layer.trainable = False
```

```
my_model.summary()
```

# Function API

```
from keras.layers import Input, Conv2D, MaxPool2D, Flatten, Dense
```

```
inp = Input(shape=(224,224,3))

conv1 = Conv2D(64, (3,3), activation='relu')(inp)

maxpool1 =  MaxPool2D()(conv1)

conv2 = Conv2D(128, (3,3), activation='relu')(maxpool1)

maxpool2 =  MaxPool2D()(conv2)

ft = Flatten()(maxpool2)

fc1 = Dense(64, activation='relu')(ft)
fc2 = Dense(4, activation='softmax')(fc1)
```

```
my_new_model = Model(inputs =  inp   , outputs =  fc2)
```

```
my_new_model.summary()
```
